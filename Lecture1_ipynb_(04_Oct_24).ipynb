### 1. Introduction to AI

**What is AI?**  
Artificial Intelligence is the field of computer science that focuses on creating machines capable of performing tasks that would normally require human intelligence. These tasks can include visual perception, speech recognition, decision-making, language translation, and much more.

**Key Characteristics of AI**:  
- **Automation**: AI systems can automate repetitive tasks and perform them efficiently.
- **Adaptation**: AI systems can improve over time by learning from data.
- **Problem-solving**: AI is designed to solve complex problems and adapt to new situations.

**Example Applications of AI**:  
- **Voice Assistants** (e.g., Siri, Alexa): AI-based systems that process natural language and respond accordingly.
- **Recommendation Systems** (e.g., Netflix, Amazon): AI systems that suggest products or content based on user behavior.
- **Self-driving Cars**: AI that uses computer vision, sensor data, and machine learning to navigate autonomously.
  
---

### 2. Types of AI

**1. Narrow AI (Weak AI)**:  
AI systems designed to perform specific tasks. They operate under a limited set of constraints and cannot perform tasks outside of their domain.  
**Example**: Face recognition software can identify faces but cannot translate languages.

**2. General AI (Strong AI)**:  
Hypothetical AI that can understand, learn, and apply intelligence to any task, just like a human. It does not yet exist.  
**Example**: A system that could perform any intellectual task a human can, such as reasoning and problem-solving across various fields.

**3. Superintelligent AI**:  
Theoretical AI that surpasses human intelligence in all respects. This level of AI remains speculative and would involve machines that could improve themselves autonomously.



---

### 3. What is Machine Learning?

**Definition**:  
Machine Learning (ML) is a subset of AI that allows machines to learn patterns from data and make decisions or predictions without being explicitly programmed for each task.

**How it Works**:  
In ML, we provide a model with data, and the model learns from the data by identifying patterns. Over time, the model improves its performance through experience. Instead of coding every rule, we let the machine figure out relationships in data.

**Key Concept**:  
- **Data**: ML models require large datasets to learn effectively.
- **Model**: A mathematical representation that identifies patterns in the data.
- **Training**: The process of feeding data to a model so it can learn.

**Example**:  
In spam email detection, we don’t code every possible rule for what constitutes spam. Instead, we train a model on a dataset containing spam and non-spam emails, allowing the model to learn patterns distinguishing the two.

---

### 4. Types of Machine Learning

**1. Supervised Learning**:  
In supervised learning, the model is trained on labeled data, where the correct output is provided. The model learns to map inputs to outputs by using this labeled data.

- **Example**: House price prediction based on features like size, location, and number of rooms (input), with the known price (output) provided.  
- **Algorithms**: Linear regression, Support Vector Machines, Decision Trees.

**2. Unsupervised Learning**:  
In unsupervised learning, the model is trained on data that has no labels. The model’s job is to find hidden patterns or relationships in the data.

- **Example**: Grouping customers based on purchasing behavior without prior knowledge of the customer categories.  
- **Algorithms**: K-means clustering, Principal Component Analysis (PCA).

**3. Reinforcement Learning**:  
In reinforcement learning, an agent interacts with an environment and learns to make decisions by receiving feedback in the form of rewards or penalties.

- **Example**: Training a robot to navigate a maze, where it receives rewards for moving closer to the goal and penalties for hitting walls.  
- **Algorithm**: Q-Learning, Deep Q-Networks (DQN).



---

### 5. What is Deep Learning?

**Definition**:  
Deep Learning is a specialized subset of machine learning that focuses on using **neural networks** with multiple layers, allowing models to learn from vast amounts of unstructured data like images, video, and text.

**Key Differences Between ML and DL**:  
- **Complexity**: Deep learning can learn from raw data (images, text) without requiring feature engineering, unlike traditional machine learning.
- **Data Requirements**: Deep learning requires large amounts of data for training, while traditional ML can work with smaller datasets.
- **Computation Power**: Deep learning models need significantly more computational power and resources (e.g., GPUs) due to their complexity.

**Neural Networks**:  
Deep learning uses artificial neural networks that mimic the way the human brain operates. Each layer of the network consists of nodes (neurons) that process data and pass the output to the next layer.

- **Input Layer**: Receives data (e.g., an image).
- **Hidden Layers**: Multiple layers that process and extract features from the data.
- **Output Layer**: Produces the final prediction or result (e.g., "cat" or "dog" for image classification).

**Example Applications**:
- **Image Recognition**: Identifying objects in images (e.g., detecting faces, cars).
- **Speech Recognition**: Converting spoken language into text

### 2. What is an Artificial Neuron?

**Definition**:  
An artificial neuron, also known as a **perceptron**, is a mathematical model inspired by biological neurons in the brain. It takes in multiple inputs, processes them, and produces a single output.

#### Structure of an Artificial Neuron:
1. **Inputs**:  
   Each neuron receives one or more inputs. These inputs can represent data features (e.g., pixels in an image, attributes of a data point).

2. **Weights**:  
   Each input is assigned a weight. The weight determines the importance of that input for the final output. If a particular input is highly relevant to the prediction, it will have a larger weight.

   \[
   z = w_1x_1 + w_2x_2 + w_3x_3 + ... + w_nx_n
   \]

   - Here, \( x_1, x_2, x_3, ... \) represent inputs.
   - \( w_1, w_2, w_3, ... \) are weights associated with each input.

3. **Bias**:  
   The bias term allows the model to shift the output to a higher or lower value independently of the input values. It improves the flexibility of the model.

   \[
   z = w_1x_1 + w_2x_2 + w_3x_3 + ... + w_nx_n + b
   \]

4. **Activation Function**:  
   After calculating the weighted sum of inputs plus the bias, the result is passed through an **activation function** to determine the neuron’s output. Activation functions introduce non-linearity into the model.

   Common activation functions:
   - **Sigmoid**: \(\sigma(z) = \frac{1}{1 + e^{-z}}\)
     - Outputs a value between 0 and 1.
   - **ReLU (Rectified Linear Unit)**: \( f(z) = \max(0, z) \)
     - Outputs the input directly if positive; otherwise, it outputs 0.

5. **Output**:  
   The output is the result of applying the activation function to the weighted sum of inputs.

**Example**:  
Let’s assume we are trying to predict whether an email is spam based on features like:
- \( x_1 \) = Number of links in the email.
- \( x_2 \) = Number of times "Free" appears in the email.
- \( x_3 \) = Length of the email.

Each input feature will have a corresponding weight, and the weighted sum is calculated. After adding bias and applying an activation function, the output will be either:
- 1 (Spam)
- 0 (Not spam)

### 3. Layers in a Neural Network

**Definition**:  
A neural network is made up of **layers of neurons**. These layers work together to process inputs and make predictions. A neural network typically consists of three types of layers:

1. **Input Layer**:  
   This layer receives the input data. Each input feature (e.g., pixel values in an image) is fed into its corresponding neuron in the input layer.

   - Example: In an image of 28x28 pixels, the input layer will have 784 neurons (one for each pixel).

2. **Hidden Layers**:  
   These layers are where the actual learning happens. Hidden layers are made up of neurons that process the inputs from the previous layer. There can be one or multiple hidden layers, and each neuron in these layers is connected to every neuron in the preceding and succeeding layer.

   - Example: For an image recognition task, hidden layers will extract patterns (e.g., edges, textures) that help in classifying the image.

3. **Output Layer**:  
   The output layer produces the final prediction. The number of neurons in the output layer depends on the task:
   - For binary classification, the output layer will have 1 neuron (0 or 1).
   - For multi-class classification, the output layer will have as many neurons as there are classes (e.g., 10 neurons for recognizing digits 0-9).

**Neural Network Example**:  
For a simple classification task (like predicting if a review is positive or negative):
- **Input Layer**: Each word in the review is converted into a numerical input.
- **Hidden Layers**: Extract patterns such as important words or phrases.
- **Output Layer**: Outputs a probability (0 = Negative, 1 = Positive).



### 4. Forward Propagation in Neural Networks

**Definition**:  
Forward propagation is the process by which data moves through the layers of a neural network, from the input layer to the output layer. In forward propagation, each neuron passes its output to the next layer, until the final prediction is made.

#### Steps in Forward Propagation:

1. **Input to Neurons**:  
   The input features (data points) are fed into the input layer neurons. Each input is multiplied by its corresponding weight, and the bias is added.

   \[
   z = w_1x_1 + w_2x_2 + w_3x_3 + ... + w_nx_n + b
   \]

2. **Activation**:  
   The weighted sum is passed through an activation function to produce the neuron’s output. This output becomes the input for the neurons in the next layer.

   \[
   a = \sigma(z)
   \]

   - If using a **ReLU** activation function:  
     \[
     a = \max(0, z)
     \]
   - If using a **Sigmoid** activation function:  
     \[
     a = \frac{1}{1 + e^{-z}}
     \]

3. **Move to Next Layer**:  
   The output from the neurons in one layer is used as the input for the neurons in the next layer. This process continues until the data reaches the output layer.

4. **Final Prediction**:  
   The output layer neurons provide the final prediction. In classification tasks, this could be a probability score for each class, and the highest probability class is chosen as the prediction.

**Example** (Handwritten Digit Classification Task):  
Suppose we have a neural network designed to classify handwritten digits (0-9):
- **Input Layer**: Receives a 28x28 pixel image as input.
- **Hidden Layers**: Process the image to recognize patterns such as edges, lines, and shapes.
- **Output Layer**: Has 10 neurons, each representing a digit (0 to 9). The neuron with the highest activation corresponds to the predicted digit.

**Mathematical Example**:  
Let’s assume:
- **Input**: A two-feature data point \( (x_1 = 0.5, x_2 = 1.0) \).
- **Weights**: For the first neuron in the hidden layer, the weights are \( w_1 = 0.8, w_2 = -0.6 \), and bias is \( b = 0.1 \).
  
Step 1: Calculate the weighted sum:
\[
z = (0.8 \times 0.5) + (-0.6 \times 1.0) + 0.1 = 0.4 - 0.6 + 0.1 = -0.1
\]

Step 2: Apply the activation function (ReLU):
\[
a = \max(0, -0.1) = 0
\]

This output (0) is passed to the next neuron in the network.

```python
# Import necessary libraries
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model

# Creating a small neural network model
model = Sequential()

# Input layer with 2 neurons (2 input features)
model.add(Dense(2, input_dim=2, activation='relu'))

# Hidden layer with 3 neurons
model.add(Dense(3, activation='relu'))

# Output layer with 1 neuron (for binary classification)
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Visualizing the model architecture
plot_file = 'simple_nn_model.png'
plot_model(model, to_file=plot_file, show_shapes=True, show_layer_names=True)

# Print model summary
model.summary()
```
